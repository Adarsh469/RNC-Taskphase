```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])
y_train = np.array([250, 300, 480,  430,   630, 730,])
```

```python
def compute_cost(x, y, w, b):
   
    m = x.shape[0] 
    cost = 0
    
    for i in range(m):
        f_wb = w * x[i] + b
        cost = cost + (f_wb - y[i])**2
    total_cost = 1 / (2 * m) * cost

    return total_cost
```

```python
def compute_gradient(x,y,w,b): #for eqn 4 and 5 and returns gradiesnt
    m = x.shape[0]
    dj_dw = 0
    dj_db = 0
    for i in range(m):
        f_wb = w*x[i] + b
        dj_dw_i = (f_wb - y[i]) * x[i]
        dj_db_i = f_wb - y[i]
        dj_dw += dj_dw_i
        dj_db += dj_db_i
    dj_dw = dj_dw/m
    dj_db = dj_db/m
    return dj_dw,dj_db
```

```python
def gradient_descent(x, y, w_init, b_init, learning_rate, iterations):
    w = w_init
    b = b_init
    for i in range(iterations):
        dj_dw, dj_db = compute_gradient(x, y, w, b)

        # Update parameters
        w = w - learning_rate * dj_dw
        b = b - learning_rate * dj_db

        if i % 100 == 0 or i == iterations - 1:
            cost = compute_cost(x, y, w, b)
            print(f"Iteration {i+1}: Cost = {cost:.4f}, w = {w:.4f}, b = {b:.4f}")

    return w, b
```

```python
w_init = 0
b_init = 0
learning_rate = 0.01
iterations = 1000

w_final, b_final = gradient_descent(x_train,y_train,w_init,b_init,learning_rate,iterations)
print(w_final)
print(b_final)
```

```python
plt.scatter(x_train,y_train, marker='x', c='r',label = "datapoints")
plt.title("housing prices")
plt.ylabel("price in 1000s of dollars")
plt.xlabel("size of house in sq feet")
plt.legend()
```

```python
plt.scatter(x_train, y_train, color = 'b', label = 'Training Data')
x_line = np.linspace(min(x_train), max(x_train), 100)
y_line = w_final * x_line + b_final
plt.plot(x_line,y_line, color = 'r', label='Regression Line')
plt.xlabel("sq feet in 1000")
plt.ylabel("cost in 1000s of dollars")
plt.title("linear regression")
plt.legend
```
